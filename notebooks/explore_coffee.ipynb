```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data
coffee = pd.read_csv("../data/crop/coffee_output.csv")
```

```python
# TODO: view first 5 rows of your data
coffee.head()
```

```python
# TODO: describe your numeric columns
coffee.describe()
```

```python
# TODO: display the null values in your dataframe
coffee.isnull().sum()
```

```python
# TODO: drop your null values
coffee.dropna(inplace=True)
```

```python
# TODO: Select only Minas Gerais
mg_coffee = coffee[coffee['subdivision'] == 'Minas Gerais']
```

```python
# TODO: BEGIN YOUR EDA!
# generate a line plot of bearing_trees changes over the years
sns.lineplot(data = mg_coffee, x = mg_coffee['year'], y = mg_coffee['bearing_trees'], markers = True, marker = 'o')
plt.xlabel('Years')
plt.ylabel('Bearing Trees')
plt.title('Bearing Trees Change Across Years')
plt.show()

# generate a Histogram with curved of bearing_trees frequency
sns.histplot(data=mg_coffee, x='bearing_trees', kde=True)
plt.title('Distribution of Bearing Trees')
plt.xlabel('Bearing Trees')
plt.ylabel('Frequency')
plt.show()

# generate a line plot of bearing_hectares changes over the years
sns.lineplot(data = mg_coffee, x = mg_coffee['year'], y = mg_coffee['bearing_hectares'], markers = True, marker = 'o')
plt.xlabel('Years')
plt.ylabel('Bearing Hectares')
plt.title('Bearing Hectares Change Across Years')
plt.show()

# generate a Histogram with curved of bearing_hectares frequency
sns.histplot(data=mg_coffee, x='bearing_hectares', kde=True)
plt.title('Distribution of Bearing Hectares')
plt.xlabel('Bearing Hectares')
plt.ylabel('Frequency')
plt.show()

# generate a line plot of 60kgs_bag changes over the years
sns.lineplot(data = mg_coffee, x = mg_coffee['year'], y = mg_coffee['60kgs_bag'], markers = True, marker = 'o')
plt.xlabel('Years')
plt.ylabel('60kgs Bag')
plt.title('60kgs Bag Change Across Years')
plt.show()

# generate a Histogram with curved of 60kgs_bag frequency
sns.histplot(data=mg_coffee, x='60kgs_bag', kde=True)
plt.title('Distribution of 60kgs Bag')
plt.xlabel('60kgs Bag')
plt.ylabel('Frequency')
plt.show()

# generate a scatter plot with bearing_trees vs. 60kgs_bag 
sns.scatterplot(data=mg_coffee, x=mg_coffee['bearing_trees'], y=mg_coffee['60kgs_bag'])
plt.title('Bearing Trees & 60kgs_bag')
plt.xlabel('Bearing Trees')
plt.ylabel('60kgs Bag')
plt.show()
# Correlation Coefficient of bearing_trees vs. 60kgs_bag 
b_and_60_corr_coef = mg_coffee['bearing_trees'].corr(mg_coffee['60kgs_bag'], method='pearson')
print('Correlation Coefficient:', b_and_60_corr_coef)

# generate a scatter plot with bearing_hectares and 60kgs_bag 
sns.scatterplot(data=mg_coffee, x=mg_coffee['bearing_hectares'], y=mg_coffee['60kgs_bag'])
plt.title('Bearing Hectares & 60kgs Bag')
plt.xlabel('Bearing Hectares')
plt.ylabel('60kgs Bag')
plt.show()
# Correlation Coefficient of bearing_hectares vs. 60kgs_bag 
bh_and_60_corr_coef = mg_coffee['bearing_hectares'].corr(mg_coffee['60kgs_bag'], method='pearson')
print('Correlation Coefficient:', bh_and_60_corr_coef) 
```
